```python
##importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
df =pd.read_csv("walmart.csv")
df.head(20)
```

```python
df.info()
```

```python
for col in df:
    print(col,": unique_values:", df[col].nunique())
```

```python
df.describe()
```

```python
df["State"].value_counts()
```

```python
orders = df.groupby("State")["Order ID"].nunique().sort_values(ascending= False)
orders
```

```python
orders = df.groupby("State")["Order ID"].nunique().sort_values(ascending= False)
#cmap1 = mcolors.LinearSegmentedColormap.from_list("mycmap", "Red")
#color1 = cmap1(np.linspace(0, 1, len(orders.index)))
plt.bar(orders.index,orders.values, color=['#00d0fa','#000000'], edgecolor='black')
plt.xticks(rotation=45)
plt.title("Number of orders by State")
for i in range(len(orders.values)):
   plt.text(i, orders.values[i]+4, orders.values[i], ha='center', fontsize=10)                            
```

```python
orders_city = df.groupby("City")["Order ID"].nunique().sort_values(ascending = False)
orders_city= orders_city[:10]
plt.bar(orders_city.index, orders_city.values, color=["red","green"], edgecolor="black")
plt.xticks(rotation=45)
plt.title("Number of orders by City")
for i in range(len(orders_city.values)):
    plt.text(i, orders_city.values[i]+4, orders_city.values[i], ha='center', fontsize=10)
```

```python
best_prod = df.groupby("Product Name").size()
best_prod = best_prod.sort_values(ascending= False)
best_prod=best_prod[:15]
plt.barh(best_prod.index,best_prod.values, color="red",edgecolor="black")
plt.title("Most bought products")
for i, v in enumerate(best_prod):
    plt.text(v+0.5,i-0.1, str(v), color='black')
```

```python
most_prod = df.groupby('Product Name')["Quantity"].mean().sort_values(ascending=False)
most_prod = most_prod[:15]
plt.barh(most_prod.index, most_prod.values, color="pink",edgecolor="black")
plt.title("avg item bought per visit (top 15) ")
for i, v in enumerate(most_prod):
    plt.text(v, i, str(v), fontsize=10)
```

```python
most_catg_item = df.groupby("Category")["Quantity"].mean().sort_values(ascending= False)
most_catg_item = most_catg_item[:15]
plt.bar(most_catg_item.index, most_catg_item.values, color="purple",edgecolor="black")
plt.title("avg amount of item bought in top 15 category")
plt.xticks(rotation=90)
for i in range(len(most_catg_item.index)):
    plt.text(i,most_catg_item.values[i]+0.05, round(most_catg_item[i],1), ha='center', fontsize=10)
```

```python
most_catg= df.groupby("Category").size().sort_values(ascending= False)
most_catg = most_catg[:8]
wedgeprops ={'linewidth':0.7, 'edgecolor':'black'}
plt.pie(most_catg, labels=most_catg.index, colors=["red","blue","purple","yellow","green","pink","brown","orange"],wedgeprops = wedgeprops, autopct='%.2f%%')
plt.title('Most bought category')
```

```python
##avg spent per order per state
state_sales= df.groupby("State")["Sales"].mean()/df.groupby("State")["Order ID"].nunique()
state_sales.sort_values(ascending= False)
state_sales=state_sales[:11]
plt.bar(state_sales.index, state_sales, color="purple", edgecolor="black")
plt.title("avg sale per order per state")
plt.xticks(rotation=45)
for i in range(len(state_sales.index)):
    plt.text(i,state_sales.values[i]+0.5,round(state_sales[i],1),ha='center', fontsize=10 )
```

```python
##avg spent per order per state without state==Wyoming
state_sales= df.groupby("State")["Sales"].mean()/df.groupby("State")["Order ID"].nunique()
state_sales.sort_values(ascending= False)
#state_sales=state_sales[:10]
state_sales= state_sales[state_sales.index!="Wyoming"]
plt.bar(state_sales.index, state_sales, color="purple", edgecolor="black")
plt.title("avg sale per order per state")
plt.xticks(rotation=45)
for i in range(len(state_sales.index)):
    plt.text(i,state_sales.values[i]+0.5,round(state_sales[i],1),ha='center', fontsize=10 )
```

```python
state_sale_total = df.groupby("State")["Sales"].sum().sort_values(ascending= False)
plt.bar(state_sale_total.index,state_sale_total, color="blue", edgecolor="black")
plt.title("statewise_sales in 10k terms")
plt.xticks(rotation=45)
for i in range(len(state_sale_total.index)):
    plt.text(i, state_sale_total.values[i]+3,round(state_sale_total[i]/10000,1),ha='center',fontsize=10)
```
